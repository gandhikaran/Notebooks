```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

```python
x=fifa.columns[54:-1]
#print(fifa.columns)
#for i in fifa.columns
   # print(i)
y=[]
y2=[]
for i in fifa.columns[54:-1]:
    temp_y=fifa.loc[fifa['Name']=='L. Messi',i].tolist()
    temp_y2=fifa.loc[fifa['Name']=='Cristiano Ronaldo',i].tolist()
    
    y.append(temp_y)
    y2.append(temp_y2)
#y=np.array(y,dtype='int16')
#np.ravel(y)
plt.figure(figsize=(20,10))
plt.plot(x,y,'.-',label='L. Messi',markersize=12)
plt.plot(x,y2,'.-',label='Cristiano Ronaldo',markersize=12)

plt.xticks(rotation=30)

temp=0
for i in fifa.columns[54:-1]:
    plt.axvline(x=temp,color='y',alpha=0.3)
    temp=temp+1

plt.title('L. Messi vs C. Ronaldo',fontdict={'fontsize':40})
plt.xlabel('Attribute',fontdict={'fontsize':20})
plt.ylabel('Rating(Out of 100)',fontdict={'fontsize':20})

plt.legend()
plt.savefig('Messi vs Ronaldo.png',dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')
#light = fifa.loc[fifa.Weight < 125].count()[0]
#light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
#medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
#medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
#heavy = fifa[fifa.Weight >= 200].count()[0]

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
rnge=np.arange(100,249,25)
weight=fifa.groupby(pd.cut(fifa["Weight"], rnge)).count().Name

#weights = [light,light_medium, medium, medium_heavy, heavy]
label = fifa.groupby(pd.cut(fifa["Weight"], rnge)).count().index
explode = (.6,.2,0,0,.4)

plt.title('Weight of the Players (lbs)')

plt.pie(weight, labels=label,explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```
