```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas
```

```python
plt.figure(figsize=(14,8))
plt.title('Gas Prices over Time',fontdict={'fontweight':'bold','fontsize':20})
plt.xlabel('Year')
plt.ylabel('Dollars per Gallon')
#plt.plot(gas.Year,gas['South Korea'],'.-',Label='South Korea')
#plt.plot(gas.Year[:10],gas.Australia[:10],'.-',Label='Australia')
#plt.plot(gas.Year,gas.Canada,'.-',Label='Canada')
#plt.plot(gas.Year,gas.USA,'.-',Label='USA')

for country in gas:
   if country != 'Year':
        plt.plot(gas.Year, gas[country],'.-',Label=country)
#plt.xticks(gas.Year[::1])        
plt.xticks(gas.Year[::1].tolist()+[2009]) #Adding extra years to the x axis
plt.legend()
plt.savefig('Gas Prices.png',dpi=200)
plt.show()

```
